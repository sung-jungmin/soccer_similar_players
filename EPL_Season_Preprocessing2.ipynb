**Data Cleaning**

```python
stats.drop(columns='index', inplace=True)
```

```python
stats.info()
```

```python
numeric_cols = [col for col in stats if stats[col].dtype.kind != 'O']
numeric_cols.remove('Appearances')
numeric_cols.remove('Tackle success %')
numeric_cols.remove('Cross accuracy %')
numeric_cols.remove('Shooting accuracy %')
len(numeric_cols)
```

5. Calculate Stats Per Match (Stats Divided by Appearances)
  --> Not: Name, Position, Appearances, Tackle success %, Cross accuracy %, Shooting accuracy%
  

```python
stats[numeric_cols]= stats[numeric_cols].div(stats['Appearances'].values,axis=0)
```

```python
stats.pop('Goals per match')
stats.pop('Passes per match')
```

```python
stats.info()
```

```python
cols = ['Season', 'Name']
stats['ID'] = stats[cols].apply(lambda row: '_'.join(row.values.astype(str)), axis=1)
```

```python
first_column = stats.pop('ID')
stats.insert(0, 'ID', first_column)
```
