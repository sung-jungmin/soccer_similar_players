```python
import numpy as np
import pandas as pd
import csv
from sklearn.metrics.pairwise import cosine_similarity
```

*Load Dataset from Google Drive*

```python
df = pd.read_csv('/content/drive/My Drive/Fall2022/EPL_Seasons/Dataset/Dataset_Exported/stats.csv',thousands=',')
```

```python
df['ID'] = df['ID'].str[:-1]
df['Name'] = df['Name'].str[:-1]
```

```python
# index
f_ID_list = df['ID'].tolist()
f_index_list = list(df.index.values)
f_idDict = dict(zip(f_index_list,f_ID_list))
```

```python
df
```

```python
df2 = df.drop(['ID','Season','Name','Position','Appearances'], axis=1)
df2
```

```python
f_v = cosine_similarity(df2.values)

cosim_df = pd.DataFrame(f_v, columns = df2.index.values, index=df2.index).reset_index()

cosim_df
```

```python
cosim_df = cosim_df.replace({'index':f_idDict})
cosim_df.rename(columns=f_idDict,inplace=True)
```

```python
cosim_df
```

```python
def sim_f_players(filename, player_league,player_season,player_name, top_num, name_dict):
  player_ID = player_league+player_season+"_"+player_name
  temporary_dataframe = pd.DataFrame(filename[player_ID].nlargest(n=top_num))
  temporary_dataframe.reset_index(inplace=True)
  final_dataframe = temporary_dataframe.replace({'index':name_dict})
  return final_dataframe
```

```python
dataframe = cosim_df
userInputLeauge = 'EPL'
userInputSeason = '20-21'
userInputName = 'Son Heung-Min' 
userInputTop = 10
nameDict = f_idDict

sim_f_players(dataframe,userInputLeauge,userInputSeason,userInputName,userInputTop,nameDict)
```

```python
league_list = ['EPL']
season_list = ['15-16','16-17','17-18','18-19','19-20','20-21']
names_list = df['Name'].unique()
df_name = pd.DataFrame(names_list, columns = ['Name'])
```

```python
#!pip install fuzzywuzzy
from fuzzywuzzy import fuzz #!pip install fuzzywuzzy if you dont own already
```

```python
def matching_score(a,b):
    return fuzz.ratio(a,b)
```

```python
def get_name_from_index(index):
   return df[df.index == index]['Name'].values[0]
```

```python
def closest_names(names):
  leven_scores = list(enumerate(df_name['Name'].apply(matching_score, b=names)))
  sorted_leven_scores = sorted(leven_scores,key=lambda x: x[1], reverse=True)
  top_closest_names = [get_name_from_index(i[0]) for i in sorted_leven_scores[:1]]
  tc_name = top_closest_names[0]
  return tc_name
```

```python
closest_names('son heung min')
```

```python
def sim_f_players_2(player_league, player_season,player_name):
  closest_name_f = closest_names(player_name)
  player_name = closest_name_f
  player_ID = player_league+player_season+"_"+player_name
  temporary_dataframe = pd.DataFrame(cosim_df[player_ID].nlargest(n=10))
  temporary_dataframe.reset_index(inplace=True)
  final_dataframe = temporary_dataframe.replace({'index':f_idDict})
  return final_dataframe
```

```python
sim_f_players_2('EPL','16-17','Son Heung-Min')
```

```python
demo = gr.Interface(
    fn = sim_f_players_2,
    inputs=[gr.Radio(league_list,label="League"),gr.Radio(season_list,label="Season"),gr.Textbox(label="Player Name")],
    outputs=[gr.outputs.Dataframe(label="Top 9 Similar Players")],
    live=True)

demo.launch()
```
