**Data Cleaning**

List of CSVs as seasons

```python
season_original = [pl1516,pl1617,pl1718,pl1819,pl1920,pl2021]
```

1. Add Season Column Then Concat

```python
season_name = ['EPL15-16','EPL16-17','EPL17-18','EPL18-19','EPL19-20','EPL20-21']
season_column = []
iteration_temp = 0
for season in season_original:
    season.insert(loc=0, column='Season', value=season_name[iteration_temp])
    season.reset_index()
    season_column.append(season)
    iteration_temp += 1
season_all = pd.concat(season_column).reset_index()
season_all.drop(columns='index')
```

2. Drop Rows with 0 Appearances

```python
season_all = season_all[season_all.Appearances != 0]
season_all.drop(columns='index')
```

3. Convert Data Type of Percentage(%) Columns Values (Object => Float)

```python
season_all['Tackle success %'] = season_all['Tackle success %'].str.rstrip('%').astype('float') / 100.0
season_all['Cross accuracy %'] = season_all['Cross accuracy %'].str.rstrip('%').astype('float') / 100.0
season_all['Shooting accuracy %'] = season_all['Shooting accuracy %'].str.rstrip('%').astype('float') / 100.0
```

4. Convert NaN/Null Values => 0

```python
season_all = season_all.replace(np.nan,0)
```
